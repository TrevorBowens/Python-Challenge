```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem
```

```python
CityDataFile="raw_data/city_data.csv"
Cities= pd.read_csv(CityDataFile)
Cities.head(10)
```

```python
RidesDataFile="raw_data/ride_data.csv"
Rides = pd.read_csv(RidesDataFile)
Rides.head()
```

```python
cityRide = pd.merge(Cities, Rides, on="city")
cityRide

```

```python
city_counts=cityRide["city"].value_counts()
city_counts.head()
```

```python
grouped_city=cityRide.groupby(["city"])

grouped_city.count().head(10)

```

```python
Pyber=pd.DataFrame()
Pyber["AverageFare"]=grouped_city["fare"].mean()
Pyber["TotalRides"]=grouped_city["ride_id"].count()
Pyber["Drivers"]=grouped_city["driver_count"].first()
Pyber["CityType"]=grouped_city["type"].first()
Pyber["TotalFares"]=grouped_city["fare"].sum()
      
Pyber.head()


      
```

```python
PyberPercentage=pd.DataFrame()
PyberPercentage["PercentageFaresByCityType"]=grouped_city["CityType"]/["TotalFares"]
PyberPercentage["PercentageRidesByCityType"]=grouped_city["TotalRides"]/["TotalFares"]
PyberPercentage["PercentageDriversByCityType"]=grouped_city["Drivers"]/["TotalFares"]

PyberPercentage.head()
```
