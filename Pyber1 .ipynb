```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from scipy.stats import sem
```

```python
CityDataFile="raw_data/city_data.csv"
Cities= pd.read_csv(CityDataFile)
Cities.head(10)
```

```python
RidesDataFile="raw_data/ride_data.csv"
Rides = pd.read_csv(RidesDataFile)
Rides.head()
```

```python
cityRide = pd.merge(Cities, Rides, on="city")
cityRide

```

```python
city_counts=cityRide["city"].value_counts()
city_counts.head()
```

```python
grouped_city=cityRide.groupby(["city"])

grouped_city.count().head(10)

```

```python
Pyber=pd.DataFrame()
Pyber["AverageFare"]=grouped_city["fare"].mean()
Pyber["TotalRides"]=grouped_city["ride_id"].count()
Pyber["Drivers"]=grouped_city["driver_count"].first()
Pyber["CityType"]=grouped_city["type"].first()
Pyber["TotalFares"]=grouped_city["fare"].sum()
      
Pyber.head(15)


      
```

```python
ICityType=Pyber.groupby('CityType')


ICityType_summary=pd.DataFrame()
ICityType_summary["TotalRides"] = ICityType["TotalRides"].sum()
ICityType_summary["Drivers"] = ICityType["Drivers"].sum()
ICityType_summary["AverageFare"] = ICityType["AverageFare"].sum()
ICityType_summary.head()

```

```python
x_axis = Pyber["TotalRides"]
y_axis = Pyber["AverageFare"] 
s = ICityType_summary["Drivers"]
plt.scatter(x_axis, y_axis, s)
plt.show
```

```python
x_limit = 50

x_axis = np.arange(0, x_limit, 5)

Shared_Rides = [Pyber["AverageFare"] for value in Pyber["TotalRides"]]

plt.scatter(x_axis, Share_Rides, marker="o", facecolor="red", edgecolors="black", s=x_axis, alpha=0.75)

plt.xlim(0, x_limit)
plt.ylim(0,40)

plt.show()

```
